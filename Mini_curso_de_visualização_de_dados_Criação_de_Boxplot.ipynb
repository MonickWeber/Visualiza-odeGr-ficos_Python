```
import matplotlib.pyplot as plt
import random #para gerar dados aleatorios

vetor = []

for i in range(100):
  num_aleatorio = random.randint(0, 50)
  vetor.append(num_aleatorio)

plt.figure(figsize=(15,8))
plt.boxplot(vetor)
plt.title('Boxplot')
plt.show()
```
