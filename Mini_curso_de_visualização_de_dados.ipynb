Inserindo um gráfico de linha bem simples

```
import matplotlib.pyplot as plt
```

```
x = [2,3]
y = [4,5]

plt.plot(x, y)

#add um titulo
plt.title("Gráfico exemplo")

#eixos

plt.xlabel('Eixo x')
plt.ylabel('Eixo y')
plt.show()
```

Criando gráfico em forma de barra

```
x = [2,3,6,9,10]
y = [4,5,6,2,8]
plt.bar(x, y, color = 'red')

#add um titulo
plt.title("Gráfico exemplo 2")

#eixos

plt.xlabel('Eixo x')
plt.ylabel('Eixo y')
plt.show()
```

Criando dois eixos e legenda

```
x1 = [2,4,6,8,10]
y1 = [4,5,6,2,8]

x2 = [1,3,5,7,9]
y2 = [2,7,1,4,5] 


plt.bar(x1, y2)
plt.bar(x2,y2)

titulo = 'Grafico de barra'
eixox = 'Eixo x'
eixoy= 'Eixo y'

#add um titulo
plt.title(titulo)

#eixos
plt.xlabel(eixox)
plt.ylabel(eixoy)


#legenda
plt.bar(x1, y1, label = 'Grupo 1')
plt.bar(x2, y2, label = 'Grupo 2')
plt.legend()
plt.show()
```

Grafico de dispersão

```
x = [2,3,4]
y = [4,5,7]

titulo = 'Gráfico de dispersão'
eixox = 'Eixo x'
eixoy = 'Eixo y'

plt.title('Gráfico de dispersão')
plt.xlabel(eixox)
plt.ylabel(eixoy)

plt.scatter(x, y, label = 'Pontos experimentais', color = 'red')
plt.plot(x, y) #plotando um grafico de linhas em cima do grafico de dispersao
plt.legend()

plt.show()
```

Brincando com o gráfico

```
x = [2,3,4]
y = [4,5,7]

titulo = 'Gráfico de dispersão'
eixox = 'Eixo x'
eixoy = 'Eixo y'

plt.title('Gráfico de dispersão')
plt.xlabel(eixox)
plt.ylabel(eixoy)

plt.scatter(x, y, label = 'Pontos experimentais', color = '#FF1493', marker='*', s=200)
plt.plot(x, y, color = 'black', linestyle = '-.') #plotando um grafico de linhas em cima do grafico de dispersao
plt.legend()

#PESQUISAR POR CORES HEXADECIMAIS E TERÁ UMA AMPLA FAIXA DE CORES

plt.show()
```

```
x = [2,3,4]
y = [4,5,7]
z = [100, 300, 400] #tamanho das estrelas no grafico

titulo = 'Gráfico de dispersão'
eixox = 'Eixo x'
eixoy = 'Eixo y'

plt.title('Gráfico de dispersão')
plt.xlabel(eixox)
plt.ylabel(eixoy)

plt.scatter(x, y, label = 'Pontos experimentais', color = '#FF1493', marker='*', s=z)
plt.plot(x, y, color = 'black', linestyle = '-.') #plotando um grafico de linhas em cima do grafico de dispersao
plt.legend()

#PESQUISAR POR CORES HEXADECIMAIS E TERÁ UMA AMPLA FAIXA DE CORES
#plt.savefig('Figura1.png', dpi = 300) #para salvar direto no pc, se salvar .pdf vai salvar em alta resolução e o dpi é o tamanho
plt.show()
```
